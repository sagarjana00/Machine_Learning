```python
#adaboost classifier

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
x, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=1)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
classifier = AdaBoostClassifier()
classifier
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [50,100,200],
              'learning_rate': [0.01,0.1,1.0,1.5,2.0],
              'algorithm': ['SAMME']
              }
```

```python
ada = AdaBoostClassifier()
clf = GridSearchCV(estimator=ada, param_grid=param_grid, cv=5, verbose=3)
clf
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
best_model = clf.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
#Adaboost Regressor

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

from sklearn.ensemble import AdaBoostRegressor
from sklearn.metrics import r2_score
```

```python
x, y = make_regression(n_samples=1000, n_features=2, noise=10, random_state=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
regressor = AdaBoostRegressor()
regressor
```

```python
regressor.fit(x_train, y_train)
```

```python
y_pred = regressor.predict(x_test)
y_pred
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [50,100,200],
              'learning_rate': [0.01,0.1, 0.5,1.0,1.5],
              'loss': ['linear', 'square', 'exponential']
              }
grid_search = GridSearchCV(estimator=AdaBoostRegressor(), param_grid=param_grid, cv=5, verbose=3)
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
y_pred_tuned = grid_search.best_estimator_.predict(x_test)
r2_score(y_pred_tuned,y_test)
```
