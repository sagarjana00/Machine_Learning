```python
# Gradient Boosting classifier and regressor

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

import warnings
warnings.filterwarnings('ignore')
```

```python
x, y = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
clf = GradientBoostingClassifier()
clf
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
#Hyper parameter Tuning

from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [50,100,200,300],
              'learning_rate': [0.01,0.1,1.0,1.5,2.0],
              }
```

```python
gbc = GradientBoostingClassifier()
grid_search = GridSearchCV(estimator=gbc, param_grid=param_grid, cv = 5, verbose=3)
```

```python
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
y_pred = grid_search.best_estimator_.predict(x_test)
```

```python
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
```

```python
#gradient boosting regressor

from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
x, y = make_regression(n_samples=1000, n_features=2, noise=10,random_state=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
regressor = GradientBoostingRegressor()
regressor
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
#hyperparameter Tuning

from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators': [50,100,200,300],
              'learning_rate': [0.01,0.1,1.0,1.5,2.0],
              }
```

```python
gbr = GradientBoostingRegressor()
grid_search = GridSearchCV(estimator=gbr, param_grid=param_grid, cv=5, verbose=3)
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
y_pred = grid_search.best_estimator_.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```
