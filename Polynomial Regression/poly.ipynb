```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#generates same synthetic datapoints or reproducibility of results
np.random.seed(1)
```

```python
#generate synthetic x
x = 2*np.random.rand(100, 1)
```

```python
x
```

```python
#generate random synthetic y using X
y = 4 + 3*x+1.5*x**2
```

```python
y
```

```python
#train test split
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_features = PolynomialFeatures(degree=2, include_bias=False)
```

```python
poly_features
```

```python
x_poly_train = poly_features.fit_transform(x_train)
x_poly_train
```

```python
x_poly_test = poly_features.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
poly_reg = LinearRegression()
poly_reg
```

```python
poly_reg.fit(x_poly_train,y_train)
```

```python
poly_reg.coef_
```

```python
poly_reg.intercept_
```

```python
#make prediction on train data itself for the exaluation matrics
y_poly_predict = poly_reg.predict(x_poly_test)
```

```python
y_poly_predict
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_poly_predict)
```

```python
plt.scatter(y_test, y_poly_predict)
plt.xlabel("Actual Values")        # label x axis
plt.ylabel("Predicted Values")     # label y axis
plt.title("Actual vs Predicted")   # add title
plt.show() 
```
