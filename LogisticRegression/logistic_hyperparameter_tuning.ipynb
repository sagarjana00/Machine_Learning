```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification
x,y = make_classification(n_samples=1000, n_features=10, n_redundant=5, n_informative=5, n_classes=2, random_state=1)
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
#Grid Search CV
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic
```

```python
param_grid = {"penalty": ('l1','l2','elasticnet'), 'C': [1, 2, 10, 20, 30, 40]}
```

```python
model = GridSearchCV(estimator=logistic, param_grid=param_grid,cv=5, verbose=2)
```

```python
model
```

```python
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
model.best_score_
```

```python
#for predict there is 2 method one is to predict directly or another way to predict building new model with the best parameters
model.best_estimator_.predict(x_test)
```

```python
#using new model

new_model = LogisticRegression(C=1, penalty='l2')
new_model
```

```python
new_model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
new_model.predict_proba(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
#Randomized search cv>> it will take randomly search CV

from sklearn.model_selection import RandomizedSearchCV
randomized_clr = RandomizedSearchCV(logistic, param_distributions=param_grid,n_iter=3, cv=5, verbose=2)
```

```python
randomized_clr
```

```python
randomized_clr.fit(x_train,y_train)
```

```python
randomized_clr.best_score_
```
