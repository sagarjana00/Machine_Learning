```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
print(data.DESCR)
```

```python
#logistic regresiion till now we have studied about 2 values of target variable(binry classification)
```

```python
data.feature_names
```

```python
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample()
```

```python
df.target.unique()
```

```python
#For making binary classification we remove the class 2

df = df[df['target'] != 2]
```

```python
df
```

```python
df.target.unique()
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.20, random_state=1)
```

```python
x_train.shape
```

```python
#model Train
from sklearn.linear_model import LogisticRegression
```

```python
classifier = LogisticRegression()
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
classifier.predict_proba(x_test)
```

```python
#evaluation metrices
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print( classification_report(y_test,y_pred))
```

```python
#macro avg>> for each class take simple avg
#support will be total no of data points 

#weight avg is use in case of class imbalance, it gives importance to performance
```

```python
from sklearn.metrics import roc_curve, auc

#get predicted probability of class 1
classifier.predict_proba(x_test)[:,1]
```

```python
y_pred_proba = classifier.predict_proba(x_test)[:,1]
y_pred_proba
```

```python
#ROC>> reciever operating characteristics curve 

fpr, tpr, thresolds = roc_curve(y_test,y_pred_proba)
```

```python
thresolds
```

```python
tpr
```

```python
fpr
```

```python
roc_auc = auc(fpr,tpr)
```

```python
plt.figure(figsize=(8, 6))
plt.plot(fpr,tpr, color='darkorange', linewidth=2, label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0,1],[0,1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('false Positive rate')
plt.ylabel('True positive rate')
plt.title('Receaver operating Characteristics curve')
plt.legend(loc='lower right')
plt.show()
```
