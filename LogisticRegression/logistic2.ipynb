```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import make_classification

x,y = make_classification(n_samples=1000, n_features=10, n_redundant=5, n_informative=5, n_classes=2, random_state=1)
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import roc_curve, auc
```

```python
y_pred_proba = model.predict_proba(x_test)[:,1]
y_pred_proba
```

```python
fpr, tpr, thresolds = roc_curve(y_test,y_pred_proba)
```

```python
#calculate auc score

roc_auc = auc(fpr, tpr)

#plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr,tpr, color='darkorange', linewidth=2, label='ROC Curve (area = %0.2f)' % roc_auc)
plt.plot([0,1],[0,1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.05])
plt.xlabel('false Positive rate')
plt.ylabel('True positive rate')
plt.title('Receaver operating Characteristics curve')
plt.legend(loc='lower right')
plt.show()
```

```python
#by default prediction class is defined with cutoff 0.5
#but if you decide your own cuttoff, you have to use precision, recall, accuracy tradeoff
#sensitivity specificity, accuracy, tradeoff
```

```python
from sklearn.metrics import precision_score, recall_score, accuracy_score

thresolds = np.linspace(0, 1, 100)
precisions = []
recalls = []
accuracies = []

for thresold in thresolds:
    y_pred_thresold = (y_pred_proba >= thresold).astype(int)
    precision = precision_score(y_test,y_pred_thresold)
    recall = recall_score(y_test,y_pred_thresold)
    accuracy = accuracy_score(y_test,y_pred_thresold)
    precisions.append(precision)
    recalls.append(recall)
    accuracies.append(accuracy)

plt.figure(figsize=(10, 6))
plt.plot(thresolds,precisions,label='Precision')
plt.plot(thresolds, recalls, label='Recall')
plt.plot(thresolds, accuracies, label='Accuracy')
plt.xlabel('Thresold Probability')
plt.ylabel('Score')
plt.title('Precision, Recall, and Accuracy vs Thresold Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# 0.4 is my optimal thresold based on precision recall accuracy trade off
```
