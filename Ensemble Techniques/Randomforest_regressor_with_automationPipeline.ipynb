```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.total_bill
```

```python
x = df.drop('total_bill', axis=1)
y = df['total_bill']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
#handling the missing value
#data encoding
#feature scaling


from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
df.sample()
```

```python
cat_cols = ["sex", "smoker", "day", "time"]
num_cols = ["tip", "size"]
```

```python
#feature engineering automation using pipeline and column transformer

num_pipeline = Pipeline(steps = [('imputation', SimpleImputer(strategy='median')),
                  ('scaling', StandardScaler()),])

cat_pipeline = Pipeline(steps = [('imputation', SimpleImputer(strategy='most_frequent')),
                  ('encoding', OneHotEncoder()),])
```

```python
preprocessor = ColumnTransformer([('num_pipeline', num_pipeline, num_cols),
                        ('cat_pipeline', cat_pipeline, cat_cols)])
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.ensemble import RandomForestRegressor
```

```python
models = {"MLR": LinearRegression(),
          "Support vector Regressor": SVR(),
          "Random forest regressor": RandomForestRegressor()}
```

```python
for i in range(len(models)):
    print(list(models.values())[i])
```

```python
from sklearn.metrics import r2_score
evaluation = {}
def model_train_eval(x_train,x_test, y_train, y_test):
    for i in range(len(models)):
        model = list(models.values())[i]
        model.fit(x_train,y_train)
        y_pred = model.predict(x_test)
        model_score = r2_score(y_test,y_pred)
        evaluation[list(models.keys())[i]] = model_score
    return evaluation
```

```python
model_train_eval(x_train,x_test,y_train,y_test)
```
