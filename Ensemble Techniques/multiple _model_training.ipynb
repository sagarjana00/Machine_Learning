```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
#predict what is the time? ... is it launch or dinner? time is the target 


#encoding, missing value treatement,scaling >> automate
```

```python
df.info()
```

```python
#Since is a nominal variable, then we will use label encoder

from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['time'] = encoder.fit_transform(df['time'])
df.head()
```

```python
df.time.unique() #dinner>>1 lunch>> 0 
```

```python
x = df.drop('time', axis=1)
y = df['time']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y , test_size=0.2, random_state=1)
```

```python
#handling the missing value
#data encoding
#feature scaling


from sklearn.impute import SimpleImputer  # for missing value
from sklearn.preprocessing import OneHotEncoder #for encoding
from sklearn.preprocessing import StandardScaler #for scaling

from sklearn.pipeline import Pipeline  #a sequence of dat transformer
from sklearn.compose import ColumnTransformer #groups all the pipeline steps for each of the columns
```

```python
df.sample()
```

```python
cat_cols = ["sex", "smoker", "day"]
num_cols = ["total_bill", "tip", "size"]
```

```python
#feature engineering automation using pipeline and column transformer

num_pipeline = Pipeline(steps = [('imputation', SimpleImputer(strategy='median')),
                  ('scaling', StandardScaler()),])

cat_pipeline = Pipeline(steps = [('imputation', SimpleImputer(strategy='most_frequent')),
                  ('encoding', OneHotEncoder()),])
```

```python
preprocessor = ColumnTransformer([('num_pipeline', num_pipeline, num_cols),
                   ('cat_pipeline', cat_pipeline, cat_cols)])
```

```python
x_train = preprocessor.fit_transform(x_train)
x_test = preprocessor.transform(x_test)
```

```python
#now data is ready, lets build the multiple model

from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
models = {"support vetor classifier": SVC(),
          "DT classifier": DecisionTreeClassifier(),
          "Log reg": LogisticRegression()}
```

```python
for i in range(len(models)):
    print(list(models.values())[i])
```

```python
from sklearn.metrics import accuracy_score

def model_train_eval(x_train,x_test,y_train,y_test):
    evaluation = {}
    for i in range(len(models)):
        model = list(models.values())[i]
        model.fit(x_train,y_train)
        y_pred = model.predict(x_test)
        model_score = accuracy_score(y_test,y_pred)
        evaluation[list(models.keys())[i]] = model_score
    return evaluation
```

```python
model_train_eval(x_train,x_test,y_train,y_test)
```
