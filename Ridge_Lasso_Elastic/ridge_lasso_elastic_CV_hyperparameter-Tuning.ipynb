```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
df = sns.load_dataset('mpg')
```

```python
df
```

```python
df.drop('name',axis=1, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
#since we have not done outlier treatement, then better idea would be replace the misssing value with median

df['horsepower'].median()
```

```python
df['horsepower']=df['horsepower'].fillna(df['horsepower'].median())
```

```python
df.isna().sum()
```

```python
df.info()
```

```python
df['origin'].value_counts()
```

```python
#data encoding
```

```python
df['origin'] = df['origin'].map({'usa':1, 'japan':2, 'europe':3})
```

```python
df.info()
```

```python
#separate into X and Y
```

```python
x = df.drop('mpg', axis=1)
y = df['mpg']
```

```python
x
```

```python
y
```

```python
#train test split
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
x_train
```

```python
x_test
```

```python
#SLR
from sklearn.linear_model import LinearRegression
regression_model = LinearRegression()
```

```python
regression_model
```

```python
regression_model.fit(x_train,y_train)
```

```python
regression_model.coef_
```

```python
for i, col_name in enumerate(x_train.columns):
    print(f"The coefficient for {col_name} is {regression_model.coef_[i]}")
```

```python
#observation 
#coefficient are relatively smaller, if one independent variable changes
#There will be not much difference in prediction
#This is sometime called as smoother model

#these features might might not be contributing in model training
```

```python
from sklearn.metrics import r2_score

y_pred_linear = regression_model.predict(x_test)
r2_linear = r2_score(y_test,y_pred_linear)
print(f"The r square of Linear regression {r2_linear}")
```

```python
#regularization
# ridge regression
```

```python
from sklearn.linear_model import Ridge
ridge_regression_model = Ridge(alpha=0.1)


#in practical implementation Lambda is alpha
```

```python
ridge_regression_model
```

```python
ridge_regression_model.fit(x_train,y_train)
```

```python
for i, col_name in enumerate(x_train.columns):
    print(f"The coefficient for {col_name} is {ridge_regression_model.coef_[i]}")
```

```python
#for ridge regression evaluation
y_pred_ridge = ridge_regression_model.predict(x_test)
r2_ridge = r2_score(y_test,y_pred_ridge)
print(f"The r square of Ridge regression {r2_ridge}")
```

```python
#we don't see much variation in coeff of ridge regression with compare to the linear regression
```

CV with Hyperparameter tuning

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.linear_model import Lasso, Ridge
```

```python
#grid search CV for Lasso
lasso = Lasso()
lasso
```

```python

param_grid = {'alpha': [0.001, 0.1, 0.01, 1,10, 100]}
param_grid
```

```python
grid_search = GridSearchCV(estimator=lasso, param_grid=param_grid, cv = 5, scoring= 'r2', verbose=2)
```

```python
grid_search
```

```python
grid_search.fit(x_train,y_train)
```

```python
grid_search.best_params_['alpha']
```

```python
grid_search.best_score_
```

```python
grid_search.best_estimator_
```

```python
y_pred  = grid_search.best_estimator_.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
#randomaize Searcg CV for lasso
```

```python
lasso = Lasso()
```

```python
param_distribution = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}
random_search_lasso = RandomizedSearchCV(estimator=lasso, param_distributions=param_distribution,n_iter=3, cv = 5, scoring='r2', verbose=2)
```

```python
random_search_lasso
```

```python
random_search_lasso.fit(x_train,y_train)
```

```python
random_search_lasso.best_score_
```

```python
y_pred  = random_search_lasso.predict(x_test)
```

```python
r2_score(y_test,y_pred)
```

```python
#Ridge Grid Search Cv
ridge = Ridge()
param_grid = {'alpha': [0.001, 0.1, 0.01, 1,10, 100]}
ridge = GridSearchCV(estimator=ridge, param_grid=param_grid, cv = 5,scoring='r2', verbose=2)
ridge.fit(x_train,y_train)
print(ridge.best_params_['alpha'])
print(ridge.best_estimator_)
y_pred = ridge.best_estimator_.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#Randomized Search Cv


ridge = Ridge()
param_distributions = {'alpha': [0.001, 0.1, 0.01, 1,10, 100]}
ridge = RandomizedSearchCV(estimator=ridge, param_distributions=param_distributions,n_iter=3, cv = 5,scoring='r2', verbose=2)
ridge.fit(x_train,y_train)
print(ridge.best_params_['alpha'])
print(ridge.best_estimator_)
y_pred = ridge.best_estimator_.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#Elastic net gridsearch CV
model = ElasticNet()
param_grid = {'alpha': [0.001, 0.1, 0.01, 1,10, 100],
              'l1_ratio': [0.1,0.4,0.9]}
model = GridSearchCV(estimator=model, param_grid=param_grid, cv = 5,scoring='r2', verbose=2)
model.fit(x_train,y_train)
print(model.best_params_['alpha'])
print(model.best_estimator_)
y_pred = model.best_estimator_.predict(x_test)
print(r2_score(y_test,y_pred))
```

```python
#Elastic net Randomized Search CV
model = ElasticNet()
param_distributions = {'alpha': [0.001, 0.1, 0.01, 1,10, 100],
              'l1_ratio': [0.1,0.4,0.9]}
model = RandomizedSearchCV(estimator=model, param_distributions=param_distributions,n_iter=3, cv = 5,scoring='r2', verbose=2)
model.fit(x_train,y_train)
print(model.best_params_['alpha'])
print(model.best_estimator_)
y_pred = model.best_estimator_.predict(x_test)
print(r2_score(y_test,y_pred))
```
