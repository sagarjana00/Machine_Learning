```python
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.linspace(-6,6,100)
y = np.sqrt(10**2-x**2)
y = np.hstack([y, -y])
x = np.hstack([x, -x])
```

```python
x
```

```python
y
```

```python
x1 = np.linspace(-6,6,100)
y1 = np.sqrt(4**2-x1**2)
y1 = np.hstack([y1, -y1])
x1 = np.hstack([x1, -x1])
```

```python
plt.scatter(y, x)
plt.scatter(y1,x1)
```

```python
df1 = pd.DataFrame(np.vstack([y,x]).T, columns = ["x1","x2"])
df1["y"] = 0

df2 = pd.DataFrame(np.vstack([y1,x1]).T, columns = ["x1","x2"])
df2["y"] = 1

df = pd.concat([df1, df2], ignore_index=True)
```

```python
df1
```

```python
df2
```

```python
df.head()
```

```python
df1 = df.copy()
df1
```

```python
df1["x1square"] = df1["x1"]**2
df1["x2square"] = df1["x2"]**2
df1["x1x2"] = df1["x1"]*df1["x2"]
df1.head()
```

```python
df1=df1.dropna()
```

```python
x = df1[["x1square", "x2square", "x1x2"]]
y = df1["y"]
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
#now without doing manually
df = df.dropna()
x = df[["x1","x2"]]
y = df["y"]

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.3, random_state=1)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
cls = SVC(kernel="poly")
cls.fit(x_train,y_train)
y_pred = cls.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#poly kernel din't perform well here 
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
cls = SVC(kernel="rbf")
cls.fit(x_train,y_train)
y_pred = cls.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
cls = SVC(kernel="sigmoid")
cls.fit(x_train,y_train)
y_pred = cls.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
#Hyperparametetuning of kernel

params = {"kernel": ["linear","rbf", "poly","sigmoid"]}

from sklearn.model_selection import GridSearchCV
model = GridSearchCV(SVC(), param_grid=params, cv=5, verbose=3)
model.fit(x_train,y_train)
```

```python
model.best_params_
```

```python
y_pred = model.best_estimator_.predict(x_test)
```

```python

accuracy_score(y_test,y_pred)
```
